```python
image = cv2.imread('/content/pollen-lowcontrast.tif', cv2.IMREAD_GRAYSCALE)
```

```python
# Apply Laplacian filter
laplacian = cv2.Laplacian(image, cv2.CV_64F)

# Apply Sobel filters to get gradient
sobelx = cv2.Sobel(image, cv2.CV_64F, 1, 0, ksize=5)
sobely = cv2.Sobel(image, cv2.CV_64F, 0, 1, ksize=5)

# Combine Sobel filters to get the gradient magnitude
gradient = np.sqrt(sobelx**2 + sobely**2)

# Display the images
plt.figure(figsize=(10, 7))
plt.subplot(1, 3, 1), plt.imshow(image, cmap='gray'), plt.title('Original')
plt.subplot(1, 3, 2), plt.imshow(laplacian, cmap='gray'), plt.title('Laplacian')
plt.subplot(1, 3, 3), plt.imshow(gradient, cmap='gray'), plt.title('Gradient')
plt.show()
```

```python
def apply_filters(image_path):
    # Load the image in grayscale
    image = cv2.imread(image_path, cv2.IMREAD_GRAYSCALE)

    # Apply Laplacian filter
    laplacian = cv2.Laplacian(image, cv2.CV_64F)

    # Apply Sobel filters to get gradient
    sobelx = cv2.Sobel(image, cv2.CV_64F, 1, 0, ksize=5)
    sobely = cv2.Sobel(image, cv2.CV_64F, 0, 1, ksize=5)

    # Combine Sobel filters to get the gradient magnitude
    gradient = np.sqrt(sobelx**2 + sobely**2)

    return image, laplacian, gradient
```

```python
image_paths = ['/content/blurry-moon.tif', '/content/blurry-moon.tif', '/content/baby.tif']
```

```python
# Set up the plot
plt.figure(figsize=(15, 10))

# Loop through the images, process, and display them
for i, path in enumerate(image_paths, 1):
    original, laplacian, gradient = apply_filters(path)

    # Plot original
    plt.subplot(3, 3, i*3-2)
    plt.imshow(original, cmap='gray')
    plt.title(f'Original {i}')

    # Plot Laplacian
    plt.subplot(3, 3, i*3-1)
    plt.imshow(laplacian, cmap='gray')
    plt.title(f'Laplacian {i}')

    # Plot Gradient
    plt.subplot(3, 3, i*3)
    plt.imshow(gradient, cmap='gray')
    plt.title(f'Gradient {i}')

plt.tight_layout()
plt.show()
```
